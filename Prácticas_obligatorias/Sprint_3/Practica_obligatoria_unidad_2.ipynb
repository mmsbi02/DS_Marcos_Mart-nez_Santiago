## PARTE II: Mini hundir la flota

```python
import numpy as np
import random

def crear_tablero(magnitud=10):
    tablero = np.full((magnitud,magnitud), " ")
    return tablero

tablero = crear_tablero()
print(tablero)
```

```python
def posicion_barcos(tablero):
    barcos =[(0,1), (1,1)],[(1,3), (1,4), (1,5), (1,6)]
    for barco in barcos:
        for (fila,columna) in barco:
            tablero[fila,columna] = "O"
    return tablero

```

```python
def disparo(tablero,fila,columna):
    if tablero[fila,columna] == "O":
        tablero[fila,columna] == "X"
        print("Tocado")
    elif tablero[fila,columna] == " ":
        tablero[fila,columna] == "-"
        print("Agua")
    else:
        print("Ahí ya disparaste")

        
    
```

```python
def barco_aleatorio(tablero, aleatorio_barco=4):
    puntos_cardinales = ["N","S","E","O"]
    magnitud = tablero.shape[0]
    colocado = False
    while not colocado:
        fila = random.randint(0, magnitud - 1)
        columna = random.randint(0, magnitud - 1)
        punto_cardinal = random.choice(puntos_cardinales)
        posiciones = []

        for i in range(aleatorio_barco):
            if punto_cardinal == "N":
                new_fila, new_columna = fila - i, columna
            elif punto_cardinal == "S":
                new_fila, new_columna = fila - i, columna
            elif punto_cardinal == "E":
                new_fila, new_columna = fila - i, columna
            else:
                new_fila, new_columna = fila - i, columna
            
            if 0 <= new_fila < magnitud and 0 <= new_columna < magnitud:
                    posiciones.append((new_fila,new_columna))
            else:
                 posiciones = []
        for f,c in posiciones:
            if posiciones and tablero[f,c] == " ":
                for f,c in posiciones:
                    tablero[f,c] ="O"
                colocado = True
    return tablero
                    
```

```python
tablero = crear_tablero()
tablero = posicion_barcos(tablero)
tablero = barco_aleatorio(tablero, 4)

print(tablero)
tablero = disparo(tablero, 0,1)
```
