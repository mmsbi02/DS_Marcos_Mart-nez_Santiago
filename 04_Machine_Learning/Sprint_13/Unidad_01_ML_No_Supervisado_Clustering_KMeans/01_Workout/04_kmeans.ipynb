#### **Centroide**

El centroide dicho brutamente es el centro de un cluster o agrupacin, es el punto promedio (es decir el que tiene como valores de sus features la media de cada feature). Aqu tienes una definicin formal (pero para luego), ahora quedate con que lo que tenemos es que dar el "centro" de los manchurrones o clusters sintticos que queremos crear.


**Definicin Formal:**
En geometra, el centroide o baricentro de una figura plana o un slido es el punto en el que se podra considerar que est concentrada toda la masa de la figura para propsitos de clculos de balance y momento. Matemticamente, el centroide de un conjunto de puntos $(x_i, y_i, z_i)$ (piensa en x,y,z como las features) en un espacio tridimensional se calcula como el promedio aritmtico de todas las coordenadas de los puntos. En trminos de frmula, si tenemos $n$ puntos, el centroide $(C_x, C_y, C_z)$ se calcula como:


$$C_x = \frac{1}{n}\sum_{i=1}^{n} x_i, \quad C_y = \frac{1}{n}\sum_{i=1}^{n} y_i, \quad C_z = \frac{1}{n}\sum_{i=1}^{n} z_i$$

En el contexto del aprendizaje automtico y el anlisis de clusters, un centroide es el punto que representa el centro de un cluster, calculado como el promedio de todas las observaciones (puntos de datos) en ese cluster. En algoritmos como K-means, los centroides son claves para determinar la pertenencia de los puntos de datos a diferentes grupos o clusters.

### Lmites de decisin

Vamos a trazar los lmites de decisin del modelo. Esto nos dar los que se denomina un *diagrama de Voronoi*. Vamos a utilizar las siguientes funciones que no son fundamentales para tu aprendizaje.

```python
def plot_data(X):
    plt.plot(X[:, 0], X[:, 1], 'k.', markersize=2)

def plot_centroids(centroids, weights=None, circle_color='w', cross_color='b'):
    if weights is not None:
        centroids = centroids[weights > weights.max() / 10]
    plt.scatter(centroids[:, 0], centroids[:, 1],
                marker='o', s=30, linewidths=8,
                color=circle_color, zorder=10, alpha=0.9)
    plt.scatter(centroids[:, 0], centroids[:, 1],
                marker='x', s=15, linewidths=20,
                color=cross_color, zorder=11, alpha=1)

def plot_decision_boundaries(clusterer, X, resolution=1000, show_centroids=True,
                             show_xlabels=True, show_ylabels=True):
    mins = X.min(axis=0) - 0.1
    maxs = X.max(axis=0) + 0.1
    xx, yy = np.meshgrid(np.linspace(mins[0], maxs[0], resolution),
                         np.linspace(mins[1], maxs[1], resolution))
    Z = clusterer.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(Z, extent=(mins[0], maxs[0], mins[1], maxs[1]),
                cmap="Pastel2")
    plt.contour(Z, extent=(mins[0], maxs[0], mins[1], maxs[1]),
                linewidths=1, colors='k')
    plot_data(X)
    if show_centroids:
        plot_centroids(clusterer.cluster_centers_)

    if show_xlabels:
        plt.xlabel("$x_1$", fontsize=14)
    else:
        plt.tick_params(labelbottom=False)
    if show_ylabels:
        plt.ylabel("$x_2$", fontsize=14, rotation=0)
    else:
        plt.tick_params(labelleft=False)
```

```python
plt.figure(figsize=(8, 4))
plot_decision_boundaries(kmeans, X)
plt.show()
```

No est mal! Es probable que algunas de las instancias cercanas a los bordes hayan sido asignadas al clster incorrecto, pero en general parece bastante bueno. (Bueno aqu se refiere a que sera el clustering que ahramos nosotros a ojo, slo  que aqu el kmeans nos simplifica la tarea, una aplicacin de automatizacin interesante sin duda)

### Hard Clustering _vs_ Soft Clustering

En lugar de elegir arbitrariamente el clster ms cercano para cada instancia, lo cual se denomina _hard_clustering_, podra ser mejor medir la distancia de cada instancia a los 5 centroides y luego nosotros implementar un mecanismo de decisin alternativo. Esto es lo que hace el mtodo `transform()`, devolvernos las distancias de cada instancia a los centroides.

```python
'''
Como hard voting vs soft voting en los ensembles
'''
# Supongamos que nos llegan nuevos datos:
X_new = np.array([[0, 2], [3, 2], [-3, 3], [-3, 2.5]])
```

Hard-clustering: nos da el cluster cuyo centroide est ms cerca de la instancia

```python
kmeans.predict(X_new)
```

Soft-Clustering: nos da las distancias a cada centroide y podemos implementar un mecanismo diferente

```python
'''
La distancia de cada punto a predecir de cada cluster. Puede ser una buena manera
de reducir el dataset a k dimensiones.
'''
kmeans.transform(X_new)
```

Puedes verificar que esta es, de hecho, la distancia euclidiana entre cada instancia y cada centroide:

```python
'''
np.tile y el reshape repite X_new 4 veces, para calcular mejor los puntos vs los centros. Luego se calcula la norma,
que es la distancia eucldea
'''
np.linalg.norm(np.tile(X_new, (1, k)).reshape(-1, k, 2) - kmeans.cluster_centers_, axis=2)
```

### K-Means Algorithm

### Algoritmo K-Means

**Funcionamiento terico**

El algoritmo K-Means es uno de los algoritmos de agrupamiento ms rpidos, pero tambin uno de los ms simples:

* Primero inicializa $k$ centroides de forma aleatoria: se eligen de forma aleatoria $k$ instancias distintas del conjunto de datos y se colocan los centroides en sus ubicaciones. [Es decir se escogen k instancias y esas son los primeros centroides]
* Ahora se asignan todas las instancias al cluster que corresponda al centroide del que estn ms cerca.
* Hecha esta eleccin se recalculan los centroides de cada cluster (es decir se hace el promedio de las features de cada instancia de cada cluster y ese es el nuevo centroide) y se vuelve a empezar... es decir
* Repetir hasta la convergencia (es decir, hasta que los centroides dejen de moverse, en general esto implica que los centroides nuevos no distan ms de un valor de tolerancia del centroide anterior o bien hasta un nmero determinado de repeticiones del reclculo de centroides):
    * Asignar cada instancia al centroide ms cercano.
    * Actualizar los centroides para que sean la media de las instancias que estn asignadas a ellos.

Realmente, la clase `KMeans` aplica por defecto versin optimizada del algoritmo anterior. Para obtener el algoritmo original de K-Means (solo con fines educativos), debes establecer `init="random"` y `n_init=1`. Estos hiperparmetros se explicarn a continuacin.

'Vamos a ejecutar el algoritmo K-Means durante 1, 2 y 3 iteraciones, para ver cmo se mueven los centroides alrededor:'

```python
'''
init: varias maneras de inicializar los clusters, por defecto una ms eficiente que random. Random -> es eleccin aleatoria de los centroides.
n_init: numero de veces que se repite el algoritmo con diferentes centroides de inicio. 
        Luego se escoge la clusterizacion o la repeticin con la mejor puntuacin en una 
        mtrica que veremos un poco ms adelante, la inercia o inertia.
max_iter: numero de veces que recalcular centroides
'''
kmeans_iter1 = KMeans(n_clusters = 5, 
                      init = 'random', 
                      n_init = 1, 
                      max_iter = 1, # 
                      random_state = 24)


kmeans_iter2 = KMeans(n_clusters = 5, 
                      init = 'random', 
                      n_init = 1, 
                      max_iter = 2, 
                      random_state = 24)


kmeans_iter3 = KMeans(n_clusters = 5, 
                      init = 'random', 
                      n_init = 1, 
                      max_iter = 3, 
                      random_state = 24)


kmeans_iter1.fit(X)
kmeans_iter2.fit(X)
kmeans_iter3.fit(X)
```

Creamos tres kmeans y vemos que ocurre con los centroides (fijate que al utilizar la misma semilla, cada modelo/algoritmo escoger los mismos centroides de partida). As podremos ver los centroides tras la primera seleccin, la segunda y la tercera.

Y vamos a pintar como se mueven los centroides y por tanto las zonas de decisin:

```python
plt.figure(figsize=(10, 8))

# Pintamos los centroides del primer modelo que solo ha hecho una iteracin es decir los centroides elegidos aleatoriamente (no se han llegado a actualizar)
plt.subplot(321) # 321 -> 3 filas, 2 columnas, elemento 1 (fila 0, columna 0)
plot_data(X)
plot_centroids(kmeans_iter1.cluster_centers_, circle_color='r', cross_color='b')
plt.ylabel("$x_2$", fontsize=14, rotation=0)
plt.tick_params(labelbottom=False)
plt.title("Actualiza centroides (inicialmente se escogen aleatoriamente)", fontsize=14)

# Pintamos los boundaries, es decir es como "etiquetamos las instancias"
plt.subplot(322) # 322 -> 3 filas, 2 columnas, elemento 2 (fila 0, columna 1)
plot_decision_boundaries(kmeans_iter1, X, show_xlabels=False, show_ylabels=False)
plt.title("Etiqueta instancias", fontsize=14)

# Sobre la eleccin anterior, pintamos los centroides de la segunda iteracin es decir los centroides recalculados
plt.subplot(323)
plot_decision_boundaries(kmeans_iter1, X, show_centroids=False, show_xlabels=False)
plot_centroids(kmeans_iter2.cluster_centers_)

# pintamos las nuevas zonas de eleccin
plt.subplot(324)
plot_decision_boundaries(kmeans_iter2, X, show_xlabels=False, show_ylabels=False)

# pintamos los centroides de la tercera iteracin
plt.subplot(325)
plot_decision_boundaries(kmeans_iter2, X, show_centroids=False)
plot_centroids(kmeans_iter3.cluster_centers_)

# pintamos la clusterizacin de la tercera iteracin
plt.subplot(326)
plot_decision_boundaries(kmeans_iter3, X, show_ylabels=False)


plt.show()
```

### Variabilidad de K-Means

En el algoritmo original de K-Means, los centroides simplemente se inicializan al azar, y el algoritmo simplemente realiza una sola iteracin para mejorar gradualmente los centroides, como vimos anteriormente.

Sin embargo, un problema importante con este enfoque es que si se ejecuta K-Means varias veces (o con diferentes semillas aleatorias), puede converger a soluciones muy diferentes, como se puede ver a continuacin:

```python
'''
Ejemplo de ejecucion de kmeans con dos semillas diferentes. Debido al inicio random, converger de manera diferente
Solucionamos con la inertia. Inicializamos varias veces, y el que tenga mejor inertia o inercia que es una mtrica que veremos en la siguiente sesin.
'''
def plot_clusterer_comparison(clusterer1, clusterer2, X, title1=None, title2=None):
    clusterer1.fit(X)
    clusterer2.fit(X)

    plt.figure(figsize=(10, 3.2))

    plt.subplot(121)
    plot_decision_boundaries(clusterer1, X)
    if title1:
        plt.title(title1, fontsize=14)

    plt.subplot(122)
    plot_decision_boundaries(clusterer2, X, show_ylabels=False)
    if title2:
        plt.title(title2, fontsize=14)
```

Creamos dos clusterings que se van a ejecutar completos pero empezando con centroides diferentes y vamos a ver como agrupan cada uno por separado.

```python
kmeans_rnd_init1 = KMeans(n_clusters = 5, init = 'random', n_init=1, random_state = 42)
kmeans_rnd_init2 = KMeans(n_clusters = 5, init = 'random', n_init=1, random_state = 141)

plot_clusterer_comparison(kmeans_rnd_init1, kmeans_rnd_init2, X,
                          "Solucion 1, rs = 42", "Solucion 2 (con una inicializacin diferente, rs = 141)")


plt.show()
```

### Inercia (Inertia)

Para seleccionar el mejor modelo, necesitaremos una forma de evaluar el rendimiento de un modelo de K-Means. Desafortunadamente, la agrupacin es una tarea no supervisada, por lo que no tenemos targets y no podemos emplear algo como la precision o el recall.  

Pero al menos podemos medir la distancia entre cada instancia y su centroide y a partir de ah sacar una mtrica. Esta es la idea detrs de la mtrica de _inercia_ o *inertia* (en ingls):

```python
'''
La suma de las distancias al cuadrado de cada punto a cada centroide
Cuanto ms baja la inertia, ms cerca estn los puntos de los centroides y en teoria la agrupacin es "mejor" 
'''
kmeans_rnd_init1.inertia_
```

```python
kmeans_rnd_init2.inertia_
```

Como puedes verificar fcilmente, la inercia es la suma de las distancias al cuadrado entre cada instancia de entrenamiento y su centroide ms cercano:

```python
X_dist = kmeans_rnd_init1.transform(X) # Obtenemos la distancia de cada instancia a los centroides
sum(X_dist[np.arange(len(X_dist)), kmeans_rnd_init1.labels_]**2) # Mira que forma tan "ingeniosa" de calcular la inercia
```

Otra forma de calcularla (aplicada al segundo kmeans)

```python
inercia = 0
for indice,vector in enumerate(X): # Vamos a recorrer todas las instancias, en concreto sus clusters asignados
    clase_asignada = kmeans_rnd_init2.labels_[indice]
    distancia = np.linalg.norm(vector - kmeans_rnd_init2.cluster_centers_[clase_asignada])
    distancia_al_cuadrado = distancia**2
    inercia += distancia_al_cuadrado
print(inercia)
```

Considerando esta mtrica (cuanto mmenos inercia mejor), el primer cluster es mejor que el segundo (visualmente tambin lo parece)

Adems no vas a tener que hacer clculos, tienes el atributo `inertia_` y el mtodo `score()` devuelve la inercia negativa. Por qu negativa? Bueno, es porque el mtodo `score()` de un predictor siempre debe respetar la regla "_mientras ms grande, mejor_".

```python
# Objetivo: minimizar la inertia
kmeans_rnd_init1.score(X)
```

Muy bien y esto de qu nos sirve? Pues para aliviar el problema de tener la dependencia de empezar con centroides aleatorios.

### Encontrar el nmero ptimo de clsteres

Si la inercia es nuestra mtrica de bondad de un modelo de clustering parece claro que una forma de encontrar el $k$ ptimo sera ir probando valores y quedarnos con el que de lugar a un agrupamiento con la inercia menor.... ### Mediante las siluetas

Lo primero ser definir que es eso de la silueta y con calma que no es de digestin rpida:

**Coeficiente de silueta**, es una medida que se aplica instancia a instancia del dataset una vez ya asociada a un clster por el algoritmo. Se define (lee y escucha con tranquilidad) como:

$$(b - a)/\max(a, b)$$

donde $a$ es la distancia media a las otras instancias en el mismo clster (es la _distancia media intra-clster_), y $b$ es la _distancia media al clster ms cercano_, es decir, la distancia media a las instancias del siguiente clster ms cercano (definido como el que minimiza $b$, excluyendo el clster de la propia instancia)

**El coeficiente de silueta puede variar entre -1 y +1: un coeficiente cercano a +1 significa que la instancia est bien dentro de su propio clster y lejos de otros clsteres, mientras que un coeficiente cercano a 0 significa que est cerca de un lmite de clster, y finalmente un coeficiente cercano a -1 significa que la instancia puede haber sido asignada al clster incorrecto.**

Ojo nosotros vamos a trabajar con el promedio (es decir la suma de todos los coeficientes de silueta dividio por el nmero de instancias del dataset), que tambin se denomina "silhoutte_score" (y que siguiendo el razonamiento anterior puede ir de -1 a +1 y que cuanto ms positivo mejor)

[Mostrar a mano como se hara el clculo para un punto]

Vamos a trazar el puntaje de la silueta como una funcin de $k$:

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(X, kmeans.labels_)
```

Empleando la lista de modelos k-means para k diferentes de la sesin anterior obtenemos el score de silueta para cada uno y lo pintamos

```python
silhouette_scores = [silhouette_score(X, model.labels_)
                     for model in kmeans_per_k[1:]]
```

```python
plt.figure(figsize=(8, 3))
plt.plot(range(2, 10), silhouette_scores, "bo-")
plt.xlabel("$k$", fontsize=14)
plt.ylabel("Silhouette score", fontsize=14)
plt.axis([1.8, 8.5, 0.55, 0.7])
plt.show()
```

Como puedes ver, esta visualizacin es mucho ms rica que la anterior: en particular, aunque confirma que $k=4$ es una muy buena eleccin, tambin subraya el hecho de que $k=5$ es igualmente bueno.

An even more informative visualization is given when you plot every instance's silhouette coefficient, sorted by the cluster they are assigned to and by the value of the coefficient. This is called a _silhouette diagram_:

Un diagrama de silueta proporciona una visualizacin an ms informativa cuando se traza el coeficiente de silueta de cada instancia, ordenado por el clster al que estn asignados y por el valor del coeficiente. Esto se conoce como un _diagrama de silueta_:

```python
from sklearn.metrics import silhouette_samples
from matplotlib.ticker import FixedLocator, FixedFormatter

'''
La anchura de cada cuchillo representa el numero de muestras por cluster. Estn ordenadas
por su coeficiente de silhouette, por eso tiene esa forma de cuchillo. Cuanta ms caida tenga
indica que las muestras tienen un coeficiente mas disperso en ese cluster
Deberian estar todos los clusters por encima de la media.
Hay algunas lineas hacia la izda xq es el coeficiente negativo. Puntos asignados al cluster erroneo.
'''
plt.figure(figsize=(11, 9))

for k in (3, 4, 5, 6):
    plt.subplot(2, 2, k - 2)
    
    y_pred = kmeans_per_k[k - 1].labels_
    silhouette_coefficients = silhouette_samples(X, y_pred)

    padding = len(X) // 30
    pos = padding
    ticks = []
    for i in range(k):
        coeffs = silhouette_coefficients[y_pred == i]
        coeffs.sort()

        color = mpl.cm.Spectral(i / k)
        plt.fill_betweenx(np.arange(pos, pos + len(coeffs)), 0, coeffs,
                          facecolor=color, edgecolor=color, alpha=0.7)
        ticks.append(pos + len(coeffs) // 2)
        pos += len(coeffs) + padding

    plt.gca().yaxis.set_major_locator(FixedLocator(ticks))
    plt.gca().yaxis.set_major_formatter(FixedFormatter(range(k)))
    if k in (3, 5):
        plt.ylabel("Cluster")
    
    if k in (5, 6):
        plt.gca().set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
        plt.xlabel("Silhouette Coefficient")
    else:
        plt.tick_params(labelbottom=True)

    plt.axvline(x=silhouette_scores[k - 2], color="red", linestyle="--")
    plt.title("$k={}$".format(k), fontsize=16)


plt.show()
```

En nuestro caso, observa que en $k=4$ y $k=5$, todos los "cuchillos" (los clusters) tienen instancias que superan el coeficiente medio (el score de silueta), pero no es el caso de $k=3$ y $k=6$, refuerza que son mejores valores para k
