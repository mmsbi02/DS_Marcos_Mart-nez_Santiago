### El problema

En concreto, vamos a trabajar con un dataset de informacin de ciudadanos como el estado civil, nmero de hijos, qu gastos e ingresos tiene, etc... y el objetivo es decidir si el ciudadano va comprar o a alquilar una vivienda. 


### Cargamos datos de entrada


Como siempre, definido el problema, y supuestamente entendido, cargamos datos y damos nuestro primer vistazo

```python
df = pd.read_csv("data/comprar_alquilar.csv")

df.head(10)
```

Por lo anterior tenemos varias features numricas continuas, varias categricas (entre ellas el target) y algunas numrica discreta.

```python
print(df.columns.to_list())
```

```python
features_num = ['ingresos', 'gastos_comunes', 'pago_coche', 'gastos_otros', 'ahorros', 'vivienda']
features_cat = ['estado_civil','trabajo']
features_num_discreta = ['hijos']
target = 'comprar'
```

### Split de train y test


```python

train_set, test_set = train_test_split(df, test_size = 0.2, random_state= 42)
```

### MiniEDA

Primero la distribuicn del target:

```python
bt.pinta_distribucion_categoricas(train_set, [target], mostrar_valores= True, relativa= True)
```

Hay un desequilibrio hacia la clase 0, as que decidimos emplear "balanced_accuracy" como mtrica.

Para esta sesin no vamos a hacer seleccin de features manual, vamos a empezar utilizndolas todas, pero como sabemos que s queremos usar PCA, tendremos que escalar y para ello las variables categricas tendremos que tratarlas con cuidado. Podemos hacer un one-hot enconding o buscar una codificacin ordinal que "tenga sentido".

Para ello, analizamos las categricas literales contra el target:

```python
for col in features_cat:
    bt.plot_categorical_relationship_fin(train_set, target, col, relative_freq= True, show_values= True)
```

En el caso de "estado_civil" podramos ordenar segn la distribucin del target == 1 (comprar) es decir 0,2,1. En el caso del trabajo podramos hacer una ordenacin similar (no 0,2,1 sino ordeanda ascendentemente por impacto en 1, a mayor proporcin mayor numero de orden). Ambas dos ordenaciones ayudarn al modelo porque a ms valor ms probabilidad de compra, pero tambin es un poquito de "sobreajuste". Vamos a aplicar el ordinal a "estado_civil" y el one-hot encoder a "trabajo", aunque podra haber combinado y con tiempo lo ms interesante (que no lo ms efectivo) habra sido probar todas las posibles combinaciones de codificaciones.

```python
map_estado_civil = {
    0: 0,
    1: 2,
    2: 1
}
train_set["estado_civil"] = train_set["estado_civil"].map(map_estado_civil)
test_set["estado_civil"] = test_set["estado_civil"].map(map_estado_civil)
```

```python
train_set = pd.get_dummies(train_set, columns = ["trabajo"], dtype = int)
test_set = pd.get_dummies(test_set, columns = ["trabajo"], dtype = int)
```

### Escalado

Necesitamos escalar (porque vamos a usar la PCA, en un momentito), por simplificar nos saltamos el paso de ver si es bueno transformar alguna de las features

```python
scaler = StandardScaler()
cols = train_set.columns.to_list()
cols.remove(target)
train_set[cols] = scaler.fit_transform(train_set[cols])
test_set[cols] = scaler.transform(test_set[cols])
```

```python
train_set.hist()
plt.tight_layout()
```

### Entra la PCA

Es el momento de aplicar la PCA para luego jugar con la reduccin de features

Primero tenemos que dividir nuestros sets en X,y para no incluir la variable target en la PCA

```python
X_train = train_set.drop(target, axis = 1)
y_train = train_set[target]

X_test = test_set.drop(target, axis = 1)
y_test = test_set[target]
```

Y ahora aplicamos el modelo de PCA entrenado contra X_train, ojo y ese mismo modelo lo aplicamos al X_test pero sin reentrenarlo (ajustarlo a X_test):

```python
from sklearn.decomposition import PCA 

pca_ml = PCA()
pca_ml.fit(X_train)
X_train_pca = pca_ml.transform(X_train)
X_test_pca = pca_ml.transform(X_test)
```

Observa que los nuevos X_train_pca y X_test_pca son ahora arrays de numpy, no es problema para sklearn que maneja arrays y dataframes:

```python
X_train_pca
```

Si queremos ver el numero de componentes princpiales (el numero de features de X_train_pca):

```python
X_train_pca.shape
```

17, que, como sabamos y esperbamos, coincide con el nmero de features originales. 

***

### Analizamos la PCA

Antes de entrenar y comparar debemos escoger el nmero de componentes (nuevas features o features de X_train_pca) que queremos quedarnos. Para ello haremos el anlisis de la varianza atribuible a cada feature igual que hicimos en sesiones pasadas


### Varianza explicada por componente

```python
# Porcentaje de varianza explicada por cada componente
# ==============================================================================
print('----------------------------------------------------')
print('Porcentaje de varianza explicada por cada componente')
print('----------------------------------------------------')
print(pca_ml.explained_variance_ratio_)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6, 4))
ax.bar(
    x      = np.arange(pca_ml.n_components_) + 1,
    height = pca_ml.explained_variance_ratio_
)

for x, y in zip(np.arange(len(X_train.columns)) + 1, pca_ml.explained_variance_ratio_):
    label = round(y, 2)
    ax.annotate(
        label,
        (x,y),
        textcoords="offset points",
        xytext=(0,10),
        ha='center'
    )

ax.set_xticks(np.arange(pca_ml.n_components_) + 1)
ax.set_ylim(0, 1.1)
ax.set_title('Porcentaje de varianza explicada por cada componente')
ax.set_xlabel('Componente principal')
ax.set_ylabel('Por. varianza explicada');
```

Podemos ver que vamos a necesitar unas cuantas ya que no hay componentes "dominantes". Veamos el grafico de ratio acumulada para escoger un posible valor inicial de numero de componentes con las que entrenar

### Ratio acumulado de Varianza por la PCA

```python
# Porcentaje de varianza explicada acumulada
# ==============================================================================
prop_varianza_acum = pca_ml.explained_variance_ratio_.cumsum()
print('------------------------------------------')
print('Porcentaje de varianza explicada acumulada')
print('------------------------------------------')
print(prop_varianza_acum)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6, 4))
ax.plot(
    np.arange(len(X_train.columns)) + 1,
    prop_varianza_acum,
    marker = 'o'
)

for x, y in zip(np.arange(len(X_train.columns)) + 1, prop_varianza_acum):
    label = round(y, 2)
    ax.annotate(
        label,
        (x,y),
        textcoords="offset points",
        xytext=(0,10),
        ha='center'
    )
    
ax.set_ylim(0, 1.1)
ax.set_xticks(np.arange(pca_ml.n_components_) + 1)
ax.set_title('Porcentaje de varianza explicada acumulada')
ax.set_xlabel('Componente principal')
ax.set_ylabel('Por. varianza acumulada');
```

Viendo la grfica una regla genrica es no bajar del 85%-90% de la varianza explicada. En este caso tendremos que emplear 11 o ms componentes principales. Partiremos de 11 e iremos subiendo para que veas la evolucin.

En un ejercicio donde la PCA sea ms necesaria (50 o ms features, por ejemplo) escogeramos un rango ms limitado en funcin de la acumulacin de varianza y si no nos diera resultado iramos aumentando el nmero de componentes (en breve vers exactamente que quiere decir quedarnos con un nmero de componentes)

### Instanciar y crear modelos sin usar el dataset PCA

Para esta sesin prctica vamos a utilizar una regresin logstica, por su sencillez, puedes probar a emplear cualquier otro modelo de clasificacin. Lo que haramos en una situacin real es utilizarya directamente el train con pca. Nosotros vamos a emplear un modelo de comparacin con todas las features orginales y as ver no slo como usar la PCA sino ver un punto de comparacin con usar todas las features originales (Nota: Ojo esto no quiere decir que siempre sea as, este es un ejemplo muy concreto)

```python
from sklearn.linear_model import LogisticRegression

lr_base = LogisticRegression(max_iter = 1000)
lr_base.fit(X_train, y_train)

lr_clf = LogisticRegression()


param_grid = {
    "max_iter": [1000, 2000, 5000],
    "C": [0.1,1,10],
    "class_weight": ["balanced", None]
}

lr_grid = GridSearchCV(lr_clf,
                       param_grid= param_grid,
                       cv = 4,
                       scoring= "balanced_accuracy")

lr_grid.fit(X_train, y_train)
```

```python
print(lr_grid.best_params_)
```

Veamos como seran el modelo de comparacin y el modelo "tuneado"

```python
print("Baseline, balanced_accuracy:", np.mean(cross_val_score(lr_base, X_train, y_train, scoring = "balanced_accuracy", cv = 4)))
print("Best LR, balanced_accuracy:", lr_grid.best_score_)
```

Competidores fuertes

### Escogiendo el nmero de PCAS:


Ahora vamos a buscar el mejor compromiso numero de PCAS/mtrica (que es lo que haramos saltndonos el paso anterior)

```python
lr_pca = LogisticRegression(max_iter = 1000, class_weight= "balanced", C = 10) # Para que pelee en justas condiciones :-)
scores_pca = []
for num_PCs in range(11,18):
    print(f"Probando con {num_PCs} componentes:", sep = " ")
    X_train_pca_reduced = X_train_pca[:,:num_PCs]
    score_pca = np.mean(cross_val_score(lr_pca, X_train_pca_reduced, y_train, cv = 4, scoring = "balanced_accuracy"))
    print(score_pca)
    scores_pca.append(score_pca)


```

Graficamente:

```python
plt.plot(range(11,18),scores_pca)
plt.xlabel("Num. PCs")
plt.ylabel("Balanced_Accuracy")
```

Los resultados son interesantes por dos motivos:
1. Puede que ms PCs no necesariamente den mejor cross-val aunque no sea mucho peor (y puede ser debido al numero de folds) pero tambin a que estamos hablando de varianza no necesariamente poder informativo.
2. Que no es necesario tener todas las componentes para igualar al modelo con todas las features.

Con cuntas PCs nos hubieramos quedado? Depende. Depende de lo que negocio o nosotros consideremos una buen valor para la mtrica y un sacrificio asumible (con slo 11 PCs, estamos perdiendo en validacin 17 puntos porcentuales respecto a usar todas, eso s reduciendo en 6/17 el nmero de features necesarias)

Una regla genrica puede ser perder un 5% sobre lo mejor a cambio de ms de un 20% de features. Si lo mejor en nuestro caso son 92% un resultado de ms de 87% con menos de 13-14 features estara "bien" en trminos generales. Nos quedamos ah, ah...
